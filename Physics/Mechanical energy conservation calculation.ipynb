역학적 에너지 보존은 E = K + U를 사용하여 계산할 수 있으며, 여기서 E는 총 역학적 에너지, K는 운동 에너지, U는 퍼텐셜 에너지이다.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the pendulum parameters
g = 9.8  # acceleration due to gravity [m/s^2]
l = 1.0  # length of pendulum [m]

# Define the initial conditions
theta0 = np.pi / 4  # initial angle [rad]
omega0 = 0.0  # initial angular velocity [rad/s]

# Time array
dt = 0.01  # time step [s]
T = 20.0  # total time [s]
t = np.arange(0, T, dt)

# Arrays for storing the results
theta = np.zeros_like(t)
omega = np.zeros_like(t)
E = np.zeros_like(t)

# Set initial conditions
theta[0] = theta0
omega[0] = omega0
E[0] = 0.5 * omega0**2 + g * l * (1 - np.cos(theta0))

# Loop through time steps
for i in range(1, len(t)):
    omega[i] = omega[i-1] - g/l * np.sin(theta[i-1]) * dt
    theta[i] = theta[i-1] + omega[i] * dt
    E[i] = 0.5 * omega[i]**2 + g * l * (1 - np.cos(theta[i]))

# Plot the results
plt.plot(t, E)
plt.xlabel('Time [s]')
plt.ylabel('Mechanical Energy [J]')
plt.show()
```

 코드에서, 우리는 중력 g로 인한 가속도와 진자의 길이를 포함하여 단순한 진자에 대한 매개 변수를 정의한다. 그런 다음 각도 세타0 및 각 속도 오메가0에 대한 초기 조건을 정의합니다. 시간 배열 t는 numpy 라이브러리의 범위 함수를 사용하여 시간 단계 dt와 총 시간 T로 생성됩니다. 그런 다음 결과를 저장하기 위해 배열 세타, 오메가 및 E를 생성하고 초기 조건을 설정합니다.

다음으로, 우리는 시간 단계를 반복하고 각 속도 오메가와 각도 세타를 업데이트하기 위해 간단한 진자에 대한 운동 방정식을 사용한다. 그런 다음 기계적 에너지 E는 방정식 E = 0.5 * 오메가^2 + g * l * (1 - np.cos(세타))를 사용하여 계산됩니다. 마지막으로, 결과는 x축의 시간 t와 y축의 기계적 에너지 E를 사용하여 matplotlib 라이브러리를 사용하여 표시됩니다.
